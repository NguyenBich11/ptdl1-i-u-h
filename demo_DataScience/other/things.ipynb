```python
import pandas as pd

# Dữ liệu ví dụ
data = {
    'Skill': ['English', 'Communicate', 'Microsoft Office', 'Lecture', 'Working Group'],
    'Level': [80,85,85,90,90],
}

# Tạo DataFrame
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
colors = ['red', 'green', 'blue', 'orange','purple']
# Vẽ biểu đồ bar
plt.figure(figsize=(8, 6))
sns.barplot(x='Skill', y='Level', data=df, width=0.6, alpha=0.8)

# Tiêu đề và nhãn trục
plt.title('My Skills')
plt.xlabel('Skill')
plt.ylabel('Level')

# Hiển thị biểu đồ
plt.show()
#plt.savefig('MySkill.jpg')
```
