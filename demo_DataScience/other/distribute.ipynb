```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm
```

```python
mu, sigma=0,0.1
s=np.random.normal(mu,sigma,1000)
count, bins, ignored=plt.hist(s,30,density=True)
plt.plot(bins,1/(sigma*np.sqrt(2*np.pi))*np.exp(-(bins-mu)**2 / (2* sigma**2)),linewidth=2,color='r')
plt.show()
```

```python
# Mean of the distribution 
Mean = 100
# satndard deviation of the distribution
Standard_deviation  = 5
# size
size = 100000
# creating a normal distribution data
values = np.random.normal(Mean, Standard_deviation, size)
# plotting histograph
plt.hist(values, 100)
# plotting mean line
plt.axvline(values.mean(), color='k', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# Tạo dữ liệu mẫu
# Phân phối lệch phải
right_skewed_data = np.random.gamma(2, 2, 1000)
# Phân phối lệch trái
left_skewed_data = np.random.beta(2, 5, 1000)
# Phân phối chuẩn
normal_data = np.random.normal(0, 1, 1000)
```

```python
# Tạo subplot
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
# Vẽ biểu đồ phân phối lệch phải
ax1.hist(right_skewed_data, bins=30, edgecolor='black', facecolor='lightgray')
ax1.set_title('Phân phối lệch phải')
ax1.set_xlabel('Giá trị')
ax1.set_ylabel('Tần suất')
# Vẽ biểu đồ phân phối lệch trái
ax2.hist(left_skewed_data, bins=30, edgecolor='black', facecolor='lightgray')
ax2.set_title('Phân phối lệch trái')
ax2.set_xlabel('Giá trị')
ax2.set_ylabel('Tần suất')
# Vẽ biểu đồ phân phối chuẩn
ax3.hist(normal_data, bins=30, edgecolor='black', facecolor='lightgray')
ax3.set_title('Phân phối chuẩn')
ax3.set_xlabel('Giá trị')
ax3.set_ylabel('Tần suất')
# Điều chỉnh spacing giữa các subplot
plt.subplots_adjust(wspace=0.4)
# Hiển thị biểu đ
```

```python
fig, ax = plt.subplots(1, 1)
a = 4
mean, var, skew, kurt = skewnorm.stats(a, moments='mvsk')
x = np.linspace(skewnorm.ppf(0.01, a),
                skewnorm.ppf(0.99, a), 100)
ax.plot(x, skewnorm.pdf(x, a),
       'r-', lw=5, alpha=0.6, label='skewnorm pdf')
vals = skewnorm.ppf([0.001, 0.5, 0.999], a)
np.allclose([0.001, 0.5, 0.999], skewnorm.cdf(vals, a))
r = skewnorm.rvs(a, size=1000)
ax.hist(r, density=True, bins='auto', histtype='stepfilled', alpha=0.2)
ax.set_xlim([x[0], x[-1]])
ax.legend(loc='best', frameon=False)
plt.show()
```
