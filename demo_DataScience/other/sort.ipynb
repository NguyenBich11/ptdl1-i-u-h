```python
import time
import random
import matplotlib.pyplot as plt
```

```python
def get_interchange_sort_complexity(arr):
    comparisons = 0
    swaps = 0
    n = len(arr)
    for i in range(n - 1):
        for j in range(i + 1, n):
            comparisons += 1
            if arr[i] > arr[j]:
                arr[i], arr[j] = arr[j], arr[i]
                swaps += 1
    return comparisons, swaps
```

```python
# Thử nghiệm với các kích thước dãy khác nhau
array_sizes = [10,20,30,40]
# Tính toán độ phức tạp Big O và vẽ biểu đồ
x_values = []
y_values = []
for size in array_sizes:
    array = random.sample(range(size), size)
    start_time = time.time()
    comparisons, swaps = get_interchange_sort_complexity(array)
    end_time = time.time()
    execution_time = end_time - start_time
    x_values.append(size)
    y_values.append(comparisons)
```

```python
# Xuất biểu đồ độ phức tạp Big O
plt.plot(x_values, y_values)
plt.xlabel('Kích thước dãy')
plt.ylabel('Số lần so sánh')
plt.title('Độ phức tạp Big O của Interchange Sort')
plt.show()
```

```python
def bubble_sort(arr):
    comparisons = 0
    swaps = 0
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            comparisons += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1
    return comparisons, swaps

def plot_bubble_sort_complexity(array_sizes):
    x_values = []
    y_values = []
    for size in array_sizes:
        array = random.sample(range(size), size)
        start_time = time.time()
        comparisons, swaps = bubble_sort(array)
        end_time = time.time()
        execution_time = end_time - start_time
        x_values.append(size)
        y_values.append(comparisons)

    plt.plot(x_values, y_values)
    plt.xlabel('Kích thước dãy')
    plt.ylabel('Số lần so sánh')
    plt.title('Độ phức tạp Big O của Bubble Sort')
    plt.show()
```

```python
# Gọi hàm để vẽ biểu đồ
plot_bubble_sort_complexity(array_sizes)
```

```python
def insertion_sort(arr):
    comparisons = 0
    shifts = 0
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        comparisons += 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            shifts += 1
            j -= 1
        arr[j + 1] = key
    return comparisons, shifts
def plot_insertion_sort_complexity(array_sizes):
    x_values = []
    y_values = []
    for size in array_sizes:
        array = random.sample(range(size), size)
        start_time = time.time()
        comparisons, shifts = insertion_sort(array)
        end_time = time.time()
        execution_time = end_time - start_time
        x_values.append(size)
        y_values.append(comparisons)
    plt.plot(x_values, y_values)
    plt.xlabel('Kích thước dãy')
    plt.ylabel('Số lần so sánh')
    plt.title('Độ phức tạp Big O của Insertion Sort')
    plt.show()
```

```python
# Gọi hàm để vẽ biểu đồ
plot_insertion_sort_complexity(array_sizes)
```

```python
def selection_sort(arr):
    comparisons = 0
    swaps = 0
    n = len(arr)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            comparisons += 1
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
        swaps += 1
    return comparisons, swaps
def plot_selection_sort_complexity(array_sizes):
    x_values = []
    y_values = []
    for size in array_sizes:
        array = random.sample(range(size), size)
        start_time = time.time()
        comparisons, swaps = selection_sort(array)
        end_time = time.time()
        execution_time = end_time - start_time
        x_values.append(size)
        y_values.append(comparisons)
    plt.plot(x_values, y_values)
    plt.xlabel('Kích thước dãy')
    plt.ylabel('Số lần so sánh')
    plt.title('Độ phức tạp Big O của Selection Sort')
    plt.show()
```

```python
# Gọi hàm để vẽ biểu đồ
plot_selection_sort_complexity(array_sizes)
```

```python
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1
def quicksort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quicksort(arr, low, pi - 1)
        quicksort(arr, pi + 1, high)
def plot_quicksort_complexity(array_sizes):
    x_values = []
    y_values = []
    for size in array_sizes:
        array = random.sample(range(size), size)
        start_time = time.time()
        quicksort(array, 0, len(array) - 1)
        end_time = time.time()
        execution_time = end_time - start_time
        x_values.append(size)
        y_values.append(execution_time)
    plt.plot(x_values, y_values)
    plt.xlabel('Kích thước dãy')
    plt.ylabel('Thời gian thực thi (giây)')
    plt.title('Độ phức tạp Big O của QuickSort')
    plt.show()
```

```python
# Gọi hàm để vẽ biểu đồ
plot_quicksort_complexity(array_sizes)
```

```python
def merge(arr, left, mid, right):
    n1 = mid - left + 1
    n2 = right - mid
    # Tạo các mảng tạm để lưu các phần tử con
    L = [0] * n1
    R = [0] * n2
    # Sao chép dữ liệu vào các mảng tạm
    for i in range(n1):
        L[i] = arr[left + i]
    for j in range(n2):
        R[j] = arr[mid + 1 + j]
    # Hợp nhất các mảng tạm vào mảng gốc
    i = 0  # Chỉ mục ban đầu của mảng con trái
    j = 0  # Chỉ mục ban đầu của mảng con phải
    k = left  # Chỉ mục ban đầu của mảng gốc
    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1
    # Sao chép các phần tử còn lại của mảng con trái (nếu có)
    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1
    # Sao chép các phần tử còn lại của mảng con phải (nếu có)
    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1
def merge_sort(arr, left, right):
    if left < right:
        mid = (left + right) // 2
        merge_sort(arr, left, mid)
        merge_sort(arr, mid + 1, right)
        merge(arr, left, mid, right)
def plot_merge_sort_complexity(array_sizes):
    x_values = []
    y_values = []
    for size in array_sizes:
        array = random.sample(range(size), size)
        start_time = time.time()
        merge_sort(array, 0, len(array) - 1)
        end_time = time.time()
        execution_time = end_time - start_time
        x_values.append(size)
        y_values.append(execution_time)
    plt.plot(x_values, y_values)
    plt.xlabel('Kích thước dãy')
    plt.ylabel('Thời gian thực thi (giây)')
    plt.title('Độ phức tạp Big O của Merge Sort')
    plt.show()
```

```python
# Gọi hàm để vẽ biểu đồ
plot_merge_sort_complexity(array_sizes)
```

```python
def bucket_sort(arr):
    n = len(arr)
    if n <= 1:
        return arr
    # Tạo các bucket rỗng
    buckets = [[] for _ in range(n)]
   # Đặt các phần tử vào bucket tương ứng
    for num in arr:
        index = int(num * n)
        buckets[index].append(num)
    # Sắp xếp từng bucket riêng biệt
    for i in range(n):
        buckets[i].sort()
    # Hợp nhất các bucket thành một mảng đã sắp xếp
    sorted_arr = []
    for bucket in buckets:
        sorted_arr.extend(bucket)

    return sorted_arr
def plot_bucket_sort_complexity(array_sizes):
    x_values = []
    y_values = []
    for size in array_sizes:
        array = [random.random() for _ in range(size)]
        start_time = time.time()
        bucket_sort(array)
        end_time = time.time()
        execution_time = end_time - start_time
        x_values.append(size)
        y_values.append(execution_time)
    plt.plot(x_values, y_values)
    plt.xlabel('Kích thước dãy')
    plt.ylabel('Thời gian thực thi (giây)')
    plt.title('Độ phức tạp Big O của Bucket Sort')
    plt.show()
```

```python
# Thử nghiệm với các kích thước dãy khác nhau
array_sizes = [100, 200,300,100]
# Gọi hàm để vẽ biểu đồ
plot_bucket_sort_complexity(array_sizes)
```
