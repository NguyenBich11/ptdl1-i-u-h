Theo quy định
Kết quả điểm sàn T1+T2+T3
Nếu >= 15 Đạt, < 15 Ko Đạt
Nếu như biết T5 T6
Tiếp tục dữ liệu: GT, KV, DT, KT Dữ liệu định tính
Câu hỏi => Học sinh có đạt điểm sàn
=> Xây dựng mô hình Decision Tree 
=> Xây dựng Decision System
KQ= f(T5,T6,GT,KV,DT,KT)
KQ --> ? {"Đạt","Không Đạt"}

```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('dulieuxettuyendaihoc.csv',encoding='utf-8',header=0,delimiter=',')
df.head(5)
```

```python
#Xử lí dữ liệu
df['DT'].fillna('K',inplace=True)
df.DT=df.DT.astype(str)
df.head(5)
```

```python
df.loc[(df['T1']+df['T2']+df['T3'])<15,'KQ']='Không Đạt'
df.loc[(df['T1']+df['T2']+df['T3'])>=15,'KQ']='Đạt'
df[['KQ']]
```

```python
df[['T1','T2','T3','T4']].corr()
```

```python
#Danh sách 
df[['T5','T6','GT','DT','KT','KQ']]
```

```python
#Encoding dữ liệu
#String --> Số
#One-hot-encoding --> Đa dạng
#Mutal
#OECoding
#Mã hóa giới tính, F là 0, M là 1
map_GT={'F':0,'M':1}
df['GT']=df['GT'].map(map_GT)
df[['GT']]
```

```python
#Kiểm tra khu vực có bao nhiêu khu vực
df.groupby(['KV']).describe()
```

```python
#Cách khác
print(np.unique(df['KV']))
```

```python
map_KV={'1':0,'2':1,'2NT':2}
df['KV']=df['KV'].map(map_KV)
df[['KV']]
```

```python
#Có bao nhiêu DT
print(np.unique(df['DT']))
```

```python
map_DT={'1.0':0,'6.0':1,'K':2}
df['DT']=df['DT'].map(map_DT)
df[['DT']]
```

```python
#Có bao nhiêu KT
print(np.unique(df['KT']))
```

```python
map_KT={'A':0,'A1':1,'B':2,'C':3,'D1':4}
df['KT']=df['KT'].map(map_KT)
df[['KT']]
```

```python
#Có bao nhiêu KQ
print(np.unique(df['KQ']))
```

```python
map_KQ={'Không Đạt':0,'Đạt':1}
df['KQ']=df['KQ'].map(map_KQ)
df[['KQ']]
```

```python
df[['T5','T6','GT','DT','KT','KQ']]
```

```python
#Phân loại Decision Tree cơ bản
#input
X=df[['T5','T6','GT','DT','KT','KQ']]
y=df.KQ
```

```python
#Chia thành 2 tập Train Test
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree=dtree.fit(X,y)
```

```python
import pydotplus
```

```python
#Xuất kết quả ra file
import pydotplus
from sklearn import tree
data_tree = tree.export_graphviz(dtree, feature_names=X.columns)
graph = pydotplus.graph_from_dot_data(data_tree)
# graph.write_jpeg('tree.jpeg')
```

```python
#Input T5=8, T6=5, GT=F, DT=K, KV=2NT,KT=A
T5=8.0
T6=5.0
GT=0
DT=2
KV=2
KT=0
kq_Predict=dtree.predict([[T5,T6,GT,DT,KV,KT]])
if(kq_Predict==0):
    print('Không Đạt')
else:
    print('Đạt')
```
