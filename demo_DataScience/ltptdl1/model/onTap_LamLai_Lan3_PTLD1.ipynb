```python
import pandas as pd
file_path = 'dulieuxettuyendaihoc.csv'
df = pd.read_csv(file_path,encoding='utf-8')
df
```

```python
median_T4=df['T4'].median()
print('Trung vị T4: ',median_T4)
```

```python
df.rename(columns={'TOANLOGICPHANTICH':'LOGIC','GIAIQUYETVANDE':'UNGXU','DINHHUONGNGHENGHIEP':'DINHHUONG'},inplace=True)
df
```

```python
df.groupby(['GT']).count()
#cho biết định tính ở OneHotEncoding
```

```python
#onehotencoding định tính giới tính
df['GT_Nam'] = df['GT'].map({'M': 1, 'F': 0})
df['GT_Nu'] = df['GT'].map({'F': 1, 'M': 0})
df
```

```python
#tạo trung bình điểm toán
df['Toan'] = df[['T1', 'T2', 'T3', 'T4', 'T5', 'T6']].mean(axis=1)
df['Toan']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split


X=df[['T1', 'T2', 'T3', 'T4', 'T5', 'T6','GT_Nam','GT_Nu']].values
y=df[['LOGIC']].values

# mức 20% random=16
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=16)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler #z-score
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import FunctionTransformer
#log , căn, bậc 2 3, sigmoid
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
import numpy as np
#poly degree=2, np.sqrt, minmaxscaler, standardscaler, LR
steps=[('polynomial', PolynomialFeatures(include_bias=False, degree=2)),('sqrttransform',FunctionTransformer(np.sqrt)),('minmaxscaler',MinMaxScaler()),('standardscaler',StandardScaler()),('model',LinearRegression())]
```

```python
pipe=Pipeline(steps)
pipe
```

```python
model=pipe.fit(X_train,y_train)
model
#chạy sai là cút
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#dự báo tập y_test y_train
y_test_pred=pipe.predict(X_test)
y_train_pred=pipe.predict(X_train)
```

```python
#tính hệ số
coefficients = model.named_steps['model'].coef_
intercept = model.named_steps['model'].intercept_
print("Coefficients góc:", coefficients)
print("Intercept tự do:", intercept)
```

```python
print('Dòng dữ liệu đầu tiên dự đoán y_train',y_train_pred[0])
```

```python
y_train[0]
```

```python
r2_score = model.score(X_train, y_train)
#biến độc lập giải thích được bao nhiêu phần trăm sự thay đổi của biến phụ thuộc
# Print the R-square score
print('R-square: ',r2_score)
```

```python
print('Value Training')
print('Score or R-Squared: ', model.score(X_train,y_train))
print('Mean Absolute Error: ',metrics.mean_absolute_error(y_train,y_train_pred))
print('Mean Squared Error: ',metrics.mean_squared_error(y_train,y_train_pred))
print('Root Mean Squared Error: ',np.sqrt(metrics.mean_squared_error(y_train,y_train_pred)))
print('Mean Absolute Percentage Error: ',metrics.mean_absolute_percentage_error(y_train,y_train_pred))
```

```python
print('Value Testing')
print('Score or R-Squared: ', model.score(X_test,y_test))
print('Mean Absolute Error: ',metrics.mean_absolute_error(y_test,y_test_pred))
print('Mean Squared Error: ',metrics.mean_squared_error(y_test,y_test_pred))
print('Root Mean Squared Error: ',np.sqrt(metrics.mean_squared_error(y_test,y_test_pred)))
print('Mean Absolute Percentage Error: ',metrics.mean_absolute_percentage_error(y_test,y_test_pred))
```
