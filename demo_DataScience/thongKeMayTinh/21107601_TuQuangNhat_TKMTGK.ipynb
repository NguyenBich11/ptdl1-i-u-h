```python
#câu 2b
print('Chinensis')
result_Chinensis=df[df['variety']=='Chinensis'].agg({
    'dimension_1_mm':['mean','median','std'],
    'dimension_2_mm':['mean','median','std'],
    'dimension_3_mm':['mean','median','std'],
})
print(result_Chinensis)
print('Peale')
result_Peale=df[df['variety']=='Peale'].agg({
    'dimension_1_mm':['mean','median','std'],
    'dimension_2_mm':['mean','median','std'],
    'dimension_3_mm':['mean','median','std'],
})
print(result_Peale)
print('Tropicalis')
result_Tropicalis=df[df['variety']=='Tropicalis'].agg({
    'dimension_1_mm':['mean','median','std'],
    'dimension_2_mm':['mean','median','std'],
    'dimension_3_mm':['mean','median','std'],
})
print(result_Tropicalis)
```
