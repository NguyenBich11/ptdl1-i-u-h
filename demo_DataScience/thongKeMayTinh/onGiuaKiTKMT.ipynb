```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Táº¡o dá»¯ liá»‡u máº«u (3D points)
data = np.random.rand(100, 3)  # Táº¡o 100 Ä‘iá»ƒm dá»¯ liá»‡u ngáº«u nhiÃªn trong khÃ´ng gian 3D

# Táº¡o biá»ƒu Ä‘á»“ hist3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Táº¡o histogram 3D
hist, xedges, yedges = np.histogram2d(data[:,0], data[:,1], bins=(20, 20))
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

# Äá»™ cao cá»§a cá»™t dá»±a trÃªn giÃ¡ trá»‹ cá»§a histogram
dz = hist.ravel()

ax.bar3d(xpos, ypos, zpos, 0.5, 0.5, dz, shade=True)

# Äáº·t tÃªn cÃ¡c trá»¥c vÃ  tiÃªu Ä‘á»
ax.set_xlabel('Trá»¥c X')
ax.set_ylabel('Trá»¥c Y')
ax.set_zlabel('Táº§n suáº¥t')
plt.title('Biá»ƒu Ä‘á»“ hist3D')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Táº¡o má»™t dÃ£y giÃ¡ trá»‹ x
x = np.linspace(-10, 10, 400)  # DÃ£y giÃ¡ trá»‹ x tá»« -10 Ä‘áº¿n 10

# TÃ­nh giÃ¡ trá»‹ y cho hai hÃ m sá»‘
y1 = 3 * x + 1
y2 = x**2 - 2*x - 3

# Váº½ Ä‘á»“ thá»‹ cá»§a hÃ m sá»‘ y1 = 3x + 1
plt.plot(x, y1, label='y = 3x + 1', color='blue')

# Váº½ Ä‘á»“ thá»‹ cá»§a hÃ m sá»‘ y2 = x^2 + 2x - 3
plt.plot(x, y2, label='y = x^2 + 2x - 3', color='red')

# TÃ¬m Ä‘iá»ƒm giao nhau báº±ng cÃ¡ch giáº£i phÆ°Æ¡ng trÃ¬nh y1 = y2
intersection_x = np.roots([1, -3, -4])  # Giáº£i phÆ°Æ¡ng trÃ¬nh x^2 + 2x - 3 - (3x + 1) = 0
intersection_y = 3 * intersection_x + 1

# Váº½ Ä‘iá»ƒm giao nhau trÃªn Ä‘á»“ thá»‹
plt.plot(intersection_x, intersection_y, 'o', color='green', label='Giao Ä‘iá»ƒm')

# Äáº·t tÃªn cho cÃ¡c trá»¥c vÃ  hiá»ƒn thá»‹ chÃº thÃ­ch
plt.xlabel('Trá»¥c X')
plt.ylabel('Trá»¥c Y')
plt.legend(loc='best')

# Hiá»ƒn thá»‹ Ä‘á»“ thá»‹
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dá»¯ liá»‡u mÃ´ phá»ng káº¿t quáº£ tung Ä‘á»“ng xu (0 lÃ  máº·t sáº¥p, 1 lÃ  máº·t ngá»­a)
data = [0, 1, 1, 0, 1, 1, 0, 0, 1, 0]

# TÃ­nh PMF báº±ng cÃ¡ch Ä‘áº¿m táº§n sá»‘ cá»§a tá»«ng giÃ¡ trá»‹
pmf = [data.count(x) / len(data) for x in set(data)]

# XÃ¡c Ä‘á»‹nh giÃ¡ trá»‹ cá»§a x vÃ  y cho biá»ƒu Ä‘á»“ PMF
x = list(set(data))
y = pmf

# Váº½ biá»ƒu Ä‘á»“ PMF
plt.bar(x, y, tick_label=x, width=0.1)
plt.xlabel('GiÃ¡ trá»‹')
plt.ylabel('XÃ¡c suáº¥t')
plt.title('PMF cá»§a phÃ¢n phá»‘i Ä‘á»“ng xu')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Äáº·t cÃ¡c tham sá»‘ cá»§a phÃ¢n phá»‘i chuáº©n (trung bÃ¬nh vÃ  Ä‘á»™ lá»‡ch chuáº©n)
mean = 0  # Trung bÃ¬nh
std_dev = 1  # Äá»™ lá»‡ch chuáº©n

# Táº¡o má»™t dÃ£y giÃ¡ trá»‹ tá»« phÃ¢n phá»‘i chuáº©n
data = np.random.normal(mean, std_dev, 1000)

# Sáº¯p xáº¿p dÃ£y giÃ¡ trá»‹
data_sorted = np.sort(data)

# TÃ­nh CDF
cdf = np.arange(1, len(data_sorted) + 1) / len(data_sorted)

# Váº½ biá»ƒu Ä‘á»“ CDF
plt.plot(data_sorted, cdf, marker='.', linestyle='none')
plt.xlabel('GiÃ¡ trá»‹')
plt.ylabel('XÃ¡c suáº¥t tÃ­ch luá»¹')
plt.title('CDF cá»§a phÃ¢n phá»‘i chuáº©n')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Äáº·t cÃ¡c tham sá»‘ cá»§a phÃ¢n phá»‘i chuáº©n (trung bÃ¬nh vÃ  Ä‘á»™ lá»‡ch chuáº©n)
mean = 0  # Trung bÃ¬nh
std_dev = 1  # Äá»™ lá»‡ch chuáº©n

# Táº¡o má»™t táº­p dá»¯ liá»‡u tá»« phÃ¢n phá»‘i chuáº©n
data = np.random.normal(mean, std_dev, 100)

# Táº¡o biá»ƒu Ä‘á»“ phÃ¢n phá»‘i chuáº©n
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', edgecolor='k')
plt.xlabel('GiÃ¡ trá»‹')
plt.ylabel('Máº­t Ä‘á»™ xÃ¡c suáº¥t')
plt.title('Histogram cá»§a phÃ¢n phá»‘i chuáº©n')
plt.grid(True)

# Táº¡o biá»ƒu Ä‘á»“ scatterplot
plt.subplot(1, 2, 2)
random_data_x = np.random.normal(mean, std_dev, 100)
random_data_y = np.random.normal(mean, std_dev, 100)
plt.scatter(random_data_x, random_data_y, alpha=0.6, c='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatterplot cá»§a phÃ¢n phá»‘i chuáº©n')
plt.grid(True)

plt.tight_layout()
plt.show()
```
