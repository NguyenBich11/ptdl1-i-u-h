```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Tạo dữ liệu mẫu (3D points)
data = np.random.rand(100, 3)  # Tạo 100 điểm dữ liệu ngẫu nhiên trong không gian 3D

# Tạo biểu đồ hist3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Tạo histogram 3D
hist, xedges, yedges = np.histogram2d(data[:,0], data[:,1], bins=(20, 20))
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

# Độ cao của cột dựa trên giá trị của histogram
dz = hist.ravel()

ax.bar3d(xpos, ypos, zpos, 0.5, 0.5, dz, shade=True)

# Đặt tên các trục và tiêu đề
ax.set_xlabel('Trục X')
ax.set_ylabel('Trục Y')
ax.set_zlabel('Tần suất')
plt.title('Biểu đồ hist3D')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo một dãy giá trị x
x = np.linspace(-10, 10, 400)  # Dãy giá trị x từ -10 đến 10

# Tính giá trị y cho hai hàm số
y1 = 3 * x + 1
y2 = x**2 - 2*x - 3

# Vẽ đồ thị của hàm số y1 = 3x + 1
plt.plot(x, y1, label='y = 3x + 1', color='blue')

# Vẽ đồ thị của hàm số y2 = x^2 + 2x - 3
plt.plot(x, y2, label='y = x^2 + 2x - 3', color='red')

# Tìm điểm giao nhau bằng cách giải phương trình y1 = y2
intersection_x = np.roots([1, -3, -4])  # Giải phương trình x^2 + 2x - 3 - (3x + 1) = 0
intersection_y = 3 * intersection_x + 1

# Vẽ điểm giao nhau trên đồ thị
plt.plot(intersection_x, intersection_y, 'o', color='green', label='Giao điểm')

# Đặt tên cho các trục và hiển thị chú thích
plt.xlabel('Trục X')
plt.ylabel('Trục Y')
plt.legend(loc='best')

# Hiển thị đồ thị
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Giả sử df là DataFrame chứa dữ liệu của bạn

# Tạo dữ liệu giả định
data = {'sex': ['MALE', 'FEMALE', 'MALE', 'FEMALE', 'MALE', 'FEMALE', 'MALE', 'FEMALE', 'MALE', 'FEMALE']}
df = pd.DataFrame(data)

# Tạo biểu đồ phân phối
sns.set(style="whitegrid")
sns.displot(data=df, x='sex', discrete=True, kde=True)
plt.xlabel('Sex')
plt.ylabel('Probability Density')
plt.title('Distribution of MALE')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dữ liệu mô phỏng kết quả tung đồng xu (0 là mặt sấp, 1 là mặt ngửa)
data = [0, 1, 1, 0, 1, 1, 0, 0, 1, 0]

# Tính PMF bằng cách đếm tần số của từng giá trị
pmf = [data.count(x) / len(data) for x in set(data)]

# Xác định giá trị của x và y cho biểu đồ PMF
x = list(set(data))
y = pmf

# Vẽ biểu đồ PMF
plt.bar(x, y, tick_label=x, width=0.1)
plt.xlabel('Giá trị')
plt.ylabel('Xác suất')
plt.title('PMF của phân phối đồng xu')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Đặt các tham số của phân phối chuẩn (trung bình và độ lệch chuẩn)
mean = 0  # Trung bình
std_dev = 1  # Độ lệch chuẩn

# Tạo một dãy giá trị từ phân phối chuẩn
data = np.random.normal(mean, std_dev, 1000)

# Sắp xếp dãy giá trị
data_sorted = np.sort(data)

# Tính CDF
cdf = np.arange(1, len(data_sorted) + 1) / len(data_sorted)

# Vẽ biểu đồ CDF
plt.plot(data_sorted, cdf, marker='.', linestyle='none')
plt.xlabel('Giá trị')
plt.ylabel('Xác suất tích luỹ')
plt.title('CDF của phân phối chuẩn')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Đặt các tham số của phân phối chuẩn (trung bình và độ lệch chuẩn)
mean = 0  # Trung bình
std_dev = 1  # Độ lệch chuẩn

# Tạo một tập dữ liệu từ phân phối chuẩn
data = np.random.normal(mean, std_dev, 100)

# Tạo biểu đồ phân phối chuẩn
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', edgecolor='k')
plt.xlabel('Giá trị')
plt.ylabel('Mật độ xác suất')
plt.title('Histogram của phân phối chuẩn')
plt.grid(True)

# Tạo biểu đồ scatterplot
plt.subplot(1, 2, 2)
random_data_x = np.random.normal(mean, std_dev, 100)
random_data_y = np.random.normal(mean, std_dev, 100)
plt.scatter(random_data_x, random_data_y, alpha=0.6, c='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatterplot của phân phối chuẩn')
plt.grid(True)

plt.tight_layout()
plt.show()
```
