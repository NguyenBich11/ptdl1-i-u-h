```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
column_name = ['Id', 'Name', 'Age', 'Weight', 'm0006', 
               'm0612', 'm1218', 'f0006', 'f0612', 'f1218']
```

```python
df = pd.read_csv('patient_heart_rate.csv')
```

```python
df.head(10)
```

```python
df[['Firstname', 'Lastname']] = df['Name'].str.split(expand=True)

df = df.drop('Name', axis = 1)
```

```python
df.head(5)
```

```python
weight = df['Weight_kgs']

for i in range(0, len(weight)):
    x = str(weight[i])
    if 'lbs' in x[-3:]: 
        x = x[:-3:]
        float_x = float(x)
        y = int(float_x/2.2)
        weight[i] = y
    if 'kgs' in x[-3:]:
        x = x[:-3:]
        float_x = float(x)
        weight[i] = x
```

```python
# Đổi tên cột Weight thành Weight_kgs
df.rename(columns={'Weight': 'Weight_kgs'}, inplace=True)
```

```python
df.head(10)
```

```python
df.dropna(how='all', inplace=True)
```

```python
df.head(10)
```

```python
df.Firstname.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
df.Lastname.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
```

```python
print(df)
```

```python
# Thay giá trị thiếu của tuổi bằng giá trị yếu vị.
df['Age'].fillna(df['Age'].mode()[0], inplace=True)
```

```python
print(df)
```

```python
# Thay giá trị thiếu của cân nặng bằng giá trị trung vị
df['Weight_kgs'].fillna(df['Weight_kgs'].median(), inplace=True)
```

```python
print(df)
```

```python
# Bước 1: Tạo melt dữ liệu để có cột gender_time
df = pd.melt(df, id_vars=['Id', 'Age', 'Weight_kgs', 'Firstname', 'Lastname'], value_name='PulseRate', var_name='gender_time').sort_values(['Id', 'Age', 'Weight_kgs', 'Firstname', 'Lastname'])
```

```python
# Bước 2: Tạo data frame tạm là kết quả của việc tách cột gender_time 
df_temp = df['gender_time'].str.extract("(\D)(\d+)(\d{2})", expand=True)
```

```python
# Bước 3: Đặt tên cột cho data frame tạm
df_temp.columns = ['Gender', 'Lower_hour', 'Upper_hour']
```

```python
# Bước 4: Nối data frame tạm vào data frame ban đầu
df = pd.concat([df, df_temp], axis=1)
```

```python
print(df)
```

```python
# Bước 5: Bỏ cột gender_time
df = df.drop(['gender_time'], axis=1)
```

```python
print (df)
```

```python
# Loại bỏ hết các dòng dữ liệu thừa là những dòng có phần PulseRate có dấu -
df = df.replace('-', np.nan).dropna(subset=['PulseRate'])
```

```python
print(df)
```

```python
# Nhận thấy có những bệnh nhân chưa ghi nhận họ tên (ví lý do nào đó)
nofirstname = df['Firstname'].isnull().sum()
nolastname = df['Lastname'].isnull().sum()

print(nofirstname)
print(nolastname)
```

```python
df['Firstname'].fillna('Unknown', inplace=True)
df['Lastname'].fillna('Unknown', inplace=True)
```

```python
df[df['Id'] == 15]
```

```python
# Sau khi xử lý thì index của dòng dữ liệu đã thay đổi lung tung, ta cần reset index lại cho 
# theo khuôn mẫu
df = df.reset_index()
```

```python
print(df)
```

```python
# Sau đó, lưu trữ dữ liệu đã xử lý thành công với tên file patient_heart_rate_clean.csv
df.to_csv('patient_heart_rate_clean.csv')
```

```python
# 2. Hãy tạo cột Fullname = Firstname+’ ‘+Lastname  
df['Fullname'] = df['Firstname'] + ' ' + df['Lastname']
```

```python
df.head(10)
```

```python
# 3. Vẽ biểu đồ cột so sánh độ tuổi giữa các tên bệnh nhân

# Tính toán độ tuổi trung bình theo tên
mean_age = df.groupby('Fullname')['Age'].mean()
```

```python
# Vẽ biểu đồ cột
plt.figure(figsize=(15, 6))
sns.barplot(x=mean_age.index, y=mean_age)
plt.title('Độ tuổi trung bình theo tên')
plt.xlabel('Tên', fontsize=12)
plt.ylabel('Độ tuổi trung bình')
plt.show()
```

```python
# 4. Vẽ biểu đồ cột so sánh nhịp tim trung bình giữa các tên bệnh nhân
# Tính toán nhịp tim trung bình theo tên
df['PulseRate'] = pd.to_numeric(df['PulseRate'])
mean_heartrate = df.groupby('Fullname')['PulseRate'].mean()
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x=mean_heartrate.index, y=mean_heartrate)
plt.title('Nhịp tim trung bình theo tên')
plt.xlabel('Tên', fontsize=12)
plt.ylabel('Nhịp tim trung bình')
plt.show()
```

```python
# 5. Vẽ biểu đồ cột so sánh nhịp tim trung bình theo từng nhóm giờ 00-06, 06-12,12-18 có 
# phân loại theo giới tính

# Tính toán nhịp tim trung bình theo giới tính và khoảng thời gian
mean_pulse = df.groupby(['Gender', 'Lower_hour', 'Upper_hour'])['PulseRate'].mean().reset_index()
```

```python
# Vẽ biểu đồ cột
sns.barplot(x='Upper_hour', y='PulseRate', hue='Gender', data=mean_pulse)
plt.title('Nhịp tim trung bình theo giới tính và khoảng thời gian')
plt.xlabel('Khoảng thời gian')
plt.ylabel('Nhịp tim trung bình (bpm)')
plt.legend(title='Giới tính')
plt.show()
```

```python
# 6. Vẽ biểu đồ phân tán thể hiện tương quan giữa trung bình độ tuổi và trung bình nhịp tim
mean_data = df.groupby('Gender').agg({'Age': 'mean', 'PulseRate': 'mean'}).reset_index()
```

```python
# Vẽ biểu đồ phân tán
sns.lmplot(data=mean_data, x='Age', y='PulseRate', hue='Gender')
plt.title('Tương quan giữa độ tuổi trung bình và nhịp tim trung bình theo giới tính')
plt.xlabel('Độ tuổi trung bình')
plt.ylabel('Nhịp tim trung bình (bpm)')
plt.show()
```

```python
# 7. Vẽ biểu đồ đường thể hiện sự thay đổi của nhịp tim trung bình các bệnh nhân theo thứ tự 
# khung giờ 00-06, 06-12,12-18 theo từng nhóm giới tính. 
mean_pulse = df.groupby(['Gender', 'Lower_hour', 'Upper_hour'])['PulseRate'].mean().reset_index()
```

```python
# Vẽ biểu đồ đường
sns.lineplot(x='Upper_hour', y='PulseRate', hue='Gender', data=mean_pulse)
plt.title('Sự thay đổi nhịp tim trung bình theo thời gian và giới tính')
plt.xlabel('Khoảng thời gian')
plt.ylabel('Nhịp tim trung bình (bpm)')
plt.legend(title='Giới tính')
plt.show()
```
