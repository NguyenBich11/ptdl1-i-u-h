```python
import numpy as np
```

```python
# 1. Sử dụng numpy để tạo một bộ dữ liệu giả lập
np.random.seed(42)
X = np.random.rand(100, 3)
y = (X[:, 0] + X[:, 1] > 1).astype(int)
```

```python
X[np.random.choice(X.shape[0], 20), np.random.choice(X.shape[1], 20)] = np.nan
```

```python
X
```

```python
y
```

```python
#  Sử dụng SimpleImputer từ sklearn.impute để điền các giá trị thiếu.
from sklearn.impute import SimpleImputer
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
imputer = SimpleImputer(strategy="mean")

X_train_imputed = imputer.fit_transform(X_train) #học xong rồi huấn luyện
X_test_imputed = imputer.transform(X_test)
```

```python
X_train_imputed
```

```python
X_test_imputed
```

```python
# 4. Chia dữ liệu thành tập huấn luyện và kiểm tra.
gnb = GaussianNB()

gnb.fit(X_train_imputed, y_train)

y_pred = gnb.predict(X_test_imputed)

# Độ chính xác
acc= accuracy_score(y_test, y_pred)

print(acc)
```
